# For

Uma lista é uma estrutura iterável, definida por um conjunto de itens indexáveis.

A isntrução range devolve um objeto iterável, constituído por uma sequência numérica com início no primeiro argumento (inclusive) e fim no segundo argumento (exludente);

Alternativamente, pode ser aapenas fornecido o número de elementos (começando a sequência em 0) out também o passo da sequência.

    for(start= , end= , step= )

start: Início da sequência

end: Final da sequência

step: passo dado em cada iteração


#### Ciclo for

```python
palavra = "Abracadabra"
cont = 0
for letra in palavra:
    if (letra == "A") or (letra == "a"):
        cont+=1
        print(f"Encontrei {cont} A")
```

```python
listaDeFruta = ["Banana","Maçã", "Laranja"]
print(f"A lista de frutas possui {len(listaDeFruta)} elemento(s)")

for fruta in listaDeFruta:
    nletras = len(fruta)
    print(f"A palavra {fruta} possui {nletras} letras")
```

```python
print("Primeiros 10 quadrados perfeitos: ")
for n in range(1,11):
    print(n**2, end=" ")
```
