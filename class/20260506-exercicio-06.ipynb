#### 3. Usando for, faça um programa que conta de 10 até 0 com range() em contagem regressiva. Utilize if para exibir uma mensagem especial quando chegar a zero. Trate possíveis erros de entrada com try/except.

```python
try:
    n = int(input("Informe um número: "))

    if(n<=0):
        print("Informe um valor maior que 0!!!")
    else:
        for i in range(n, -1, -1):
            print(f"{i}", end=" ")
            if(i==0):
                print(f"\nComo diria minha ex: Terminamos!!")

except ValueError:
    print("Informe um valor numérico!")
except TypeError:
    print("O valor deve ser um inteiro!")
```

#### 4. Usando for e if, escreva um programa que mostra os números de 1 a 20, mas salta os múltiplos de 3 com continue. Inclua try/except para lidar com entradas inválidas se o utilizador definir o intervalo.

```python
try:
    n1 = int(input("Digite o valor inicial: "))
    n2 = int(input("Digite o valor final: "))

    if(n1>n2):
        aux = n1
        n1 = n2
        n2 = aux
    elif(n1==n2):
        print("Intervalo degenerado, informe um intervalo com amplitude maior que 0.")
    else:
        for i in range(n1, n2+1, 1):
            if(i%7==0):
                print(i)
                break
except ValueError:
    print("Informe um valor numérico!")
except TypeError:
    print("Informe um valor inteiro!")
    
```

#### 7. Usando while, for, if e try/except, crie um programa que soma números inteiros positivos fornecidos pelo utilizador até que ele insira um número negativo. O programa deve pedir repetidamente novos valores até encontrar o negativo.

```python
soma = 0
while(True):
    try:
        n = int(input("Digite um valor inteiro e positivo: "))

        if n<=0:
            break
        else:
            soma += n
    except ValueError:
        print("Informe um valor numérico!")
    except TypeError:
        print("Informe um valor inteiro!")
print(soma)
```

#### 9. Usando for, if, while e try/except, crie um programa que apresenta os primeiros N quadrados perfeitos (1, 4, 9, 16, 25, …), onde N é um número natural introduzido pelo utilizador.

```python
try: 
    n = int(input("Informe um valor numérico: "))

    if(n<0):
        print("Informe um número inteiro maior que 0!")
    else:
        for i in range(1, n+1, 1):
            print(f"{i}^2 = {i**2}", end="\n")
except ValueError:
    print("Informe um valor numérico!")
```

#### 18. (Desafio) Usando for, if, while e try/except, crie um programa que verifica se um número é primo. O programa deve pedir novos números até o utilizador decidir sair.

```python
try: 
    while(True):
        n = int(input("Informe um valor numérico: "))
        if(n<0):
            break
        else:
            cont = 0
            for i in range(1, n+1, 1):
                if(n%i==0): cont+=1
            if cont == 2: print(f"{n} é primo!")
except ValueError:
    print("Informe um valor numérico!")  
```
