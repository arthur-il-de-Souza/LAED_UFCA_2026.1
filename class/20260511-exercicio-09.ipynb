#### EXERCÍCIO 2
Com a lista numeros = [1, 2, 3, 4, 5], escreva o código para adicionar o número 6 ao final da lista usando
o método append()

```python
# sem append 1

numeros = [1, 2, 3, 4, 5]
numeros = numeros + [6]

for i in numeros:
    print(i, end=" ")
```

```python
# sem append 2

numeros = [1, 2, 3, 4, 5]
aux = [0] * 6

for i in range(len(numeros)+1):
    if(i == (len(numeros))):
        aux[i] = 6
        numeros = aux
        break  
    aux[i] = numeros[i]

for i in aux:
    print(i, end=" ")
```

```python
# com append

numeros = [1, 2, 3, 4, 5]
numeros.append(6)

for i in aux:
    print(i, end=" ")
```

#### EXERCÍCIO 3
Dada a lista animais = ["gato", "cachorro", "pássaro"], adicione a lista ["peixe", "tartaruga"] à lista
animais usando o método extend().

```python
animais = ["gato", "cachorro", "pássaro"]
lista = ["peixe", "tartaruga"]

print("Animais")
for i in range(len(animais)):
    print(f"{i+1}: {animais[i]}")
print("\nLista")
for i in range(len(lista)):
    print(f"{i+1}: {lista[i]}")

animais.extend(lista)
print("\nAnimais")
for i in range(len(animais)):
    print(f"{i+1}: {animais[i]}")
```

#### EXERCÍCIO 4
Usando a lista cidades = ["Lisboa", "Porto", "Coimbra", "Faro"], insira a cidade "Braga" na terceira
posição da lista (índice 2) usando o método insert().

```python
cidades = ["Lisboa", "Porto", "Coimbra", "Faro"]
for i in range(len(cidades)):
    if i == 2:
        cidades.insert(2,"Braga")

for i in cidades:
    print(i, end=" ")
```

#### EXERCÍCIO 7
Considere a lista alfabeto = ["a", "d", "c", "b", "e"]. Ordene a lista em ordem alfabética usando o
método sort(), depois inverta a ordem dos elementos com reverse() e imprima o resultado final.

```python
alfabeto = ["a", "d", "c", "b", "e"]

print("Lista original:")
for i in alfabeto:
    print(i, end=" ")
print("\n")

print("Lista alternada:")
alfabeto.sort()
for i in alfabeto:
    print(i, end=" ")
print("\n")

print("Lista reversa")
alfabeto.reverse()
for i in alfabeto:
    print(i, end=" ")
print("\n")
```

#### EXERCÍCIO 8
Dada a lista notas = [8, 6, 9, 7, 8, 10, 8], conte quantas vezes o número 8 aparece na lista usando o
método count(). Em seguida, remova todas as ocorrências de 8 usando o método remove() em um
loop while até que 8 não esteja mais na lista.

```python
notas = [8, 6, 9, 7, 8, 10, 8]
for i in notas:
    print(i, end=" ")
print("\n")

num = notas.count(8)
print(f"O número 8 aparece {num} vez(es).")

for i in range(num):
    notas.remove(8)

for i in notas:
    print(i, end=" ")
print("\n")
```

#### EXERCÍCIO 9
Com a lista itens = [1, 2, 3, 4, 5, 6], remova todos os elementos da lista usando o método clear(). Em
seguida, insira os números 10, 20 e 30 na lista usando append() e insert(). Por fim, encontre o índice
do número 20 usando index() e imprima o resultado.

```python
itens = [1, 2, 3, 4, 5, 6]
for i in itens:
    print(i, end=" ")
print("\n")

itens.clear()

itens.append(10)
itens.append(20)
itens.append(30)


for i in itens:
    print(i, end=" ")
print("\n")

print(f"\nindex = [{itens.index(20)}]")
```
