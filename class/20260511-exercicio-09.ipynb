#### EXERCÍCIO 1
Dada a lista frutas = ["maçã", "banana", "laranja", "maçã"], qual é o comprimento da lista? Faça um
com o for e outra com a função len().

```python
# Usando for
frutas = ["maçã", "banana", "laranja", "maçã"]
cont = 0
for i in frutas:
    cont += 1
print(f"Quantidade de itens na lista: {cont}")
```

```python
frutas = ["maçã", "banana", "laranja", "maçã"]
print(f"Quantidade de itens na lista: {len(frutas)}")
```

#### EXERCÍCIO 2
Com a lista numeros = [1, 2, 3, 4, 5], escreva o código para adicionar o número 6 ao final da lista usando
o método append()

```python
# sem append

numeros = [1, 2, 3, 4, 5]
aux = [0] * 6

for i in range(len(numeros)+1):
    if(i == (len(numeros))):
        aux[i] = 6
        numeros = aux
        break  
    aux[i] = numeros[i]

for i in aux:
    print(i, end=" ")
```

```python
# com append

numeros = [1, 2, 3, 4, 5]
numeros.append(6)

for i in aux:
    print(i, end=" ")
```

#### EXERCÍCIO 3
Dada a lista animais = ["gato", "cachorro", "pássaro"], adicione a lista ["peixe", "tartaruga"] à lista
animais usando o método extend().

```python
animais = ["gato", "cachorro", "pássaro"]
lista = ["peixe", "tartaruga"]

print("Animais")
for i in range(len(animais)):
    print(f"{i+1}: {animais[i]}")
print("\nLista")
for i in range(len(lista)):
    print(f"{i+1}: {lista[i]}")

animais.extend(lista)
print("\nAnimais")
for i in range(len(animais)):
    print(f"{i+1}: {animais[i]}")
```
