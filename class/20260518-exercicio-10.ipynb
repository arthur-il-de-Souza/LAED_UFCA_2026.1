# Trabalhando com Listas Ordenadas

## 1. Introdução

### Funções:

```python
def findM(lista: list[int]):
    m=0
    for i in lista:
        if not(i==0):
            m+=1
            continue
        break
    return m-1


def baseSwape( a: int, b: int):
    aux = a
    a = b
    b = aux

    return a, b
```

### Listas:

```python
v_1 = [3, 9, 2, 3, 2, 7, 3, 9, 5, 0, 0, 0] # Repetida/Desordenada
v_2 = [2, 2, 3, 3, 3, 5, 7, 9, 9, 0, 0, 0] # Repetido/Ordenado 
v_3 = [1, 3, 7, 8, 10, 16, 19, 21, 28, 0, 0, 0]
v_4 = [3,9,10,14,18,20, 1,5,8,15,17,19] # Ordenada Intercalada

v_v = [5, 3, 9, 7, 8, 10, 13, 1, 2, 14, 21, 11]
v_u = [4, 8, 2, 11, 6, 5, 15, 3, 19, 1, 7, 19]
```

#### a) Números Repetidos em uma Lista Desordenada

```python
m = findM(v_1)
print(f"pos: {m}, num: {v_1[m]}")

for i in range(0, len(v_1), 1):
    for j in range(i+1, len(v_1), 1):
        if(v_1[i] == v_1[j]):
            v_1[j] = v_1[m+1]

m = findM(v_1)
print(f"pos: {m}, num: {v_1[m]}")

for i in range(0, len(v_1)-1, 1):
    m = findM(v_1)
    m+=1
    for j in range(m, len(v_1), 1):
        if v_1[j]==0:
            continue
        v_1[m], v_1[j] = baseSwape(v_1[m], v_1[j])
        break

for i in range(len(v_1)):
    print(f"{i}: {v_1[i]}")
        
```

#### b) Números Repetidos em uma Lista Ordenada

```python
print(v_2)

m = findM(v_2)
for i in range(1, len(v_2), 1):
    if v_2[i]==v_2[i-1]:
        v_2[i-1]=v_2[m+1]
print(v_2)

for i in range(0, len(v_2), 1):
    m = findM(v_2) + 1
    if v_2[i] != 0:
        v_2[i], v_2[m] = baseSwape(v_2[i], v_2[m])

print(v_2)
```

### Exemplos: 

#### a) Os elementos comuns Listas Desordenadas

```python
print(v_v)
print(v_u)

for i in range(0, len(v_v), 1):
    for j in range(0, len(v_u), 1):
        if v_v[i] == v_u[j]:
            print(v_v[i]) 
            break
```

#### a) Os Elementos Comuns Listas Ordenadas

```python
v_v.sort()
v_u.sort()

print(v_v)
print(v_u)

j = 0
i = 0
while(j<len(v_v) and i<len(v_u)):
    if(v_v[i] == v_u[j]):
        print(v_v[i])
        i+=1
        j+=1
    elif(v_v[i]<v_u[j]):
        i+=1
    else:
        j+=1
```

#### b) Operação de Inserção Lista Ordenada

```python
m = findM(v_3)
print(v_3)

k = 1

i = 0
while((v_3[i]<=k) and not(v_3[i]==0)):    
    i+=1

for j in range(m+1,0,-1):
    if j == i:
        v_3[i] = k
        m+=1
        break
    v_3[j], v_3[j-1] = baseSwape(v_3[j], v_3[j-1])

print(v_3)
```

```python
m = findM(v_3)
print(v_3)

k = 90

for j in range(m, -1, -1):
    if(v_3[j]>k):
        v_3[j+1], v_3[j] = baseSwape(v_3[j+1], v_3[j])
    elif(m<len(v_3)-1):
        m+=1
        v_3[j+1] = k
        break
print(v_3)
    
```

#### c) Operação de Remoção Lista Ordenada

```python
m = findM(v_3)
print(m)
print(v_3)

k = 10
i = 0
for j in range(0, m+1, 1):
    if(v_3[j] == k):
        v_3[j] = 0
        break
    i+=1
for j in range(i, m+1, 1):
    v_3[j+1], v_3[j] = baseSwape(v_3[j+1], v_3[j])
print(v_3)
```

#### d) Atualização em uma Lista Ordenada

```python
m = findM(v_3)
print(v_3)

k = 3
numero = 20

for i in range(0, m+1, 1):
    if(v_3[i] == k):
        v_3[i] = numero
        break
print(v_3)

if v_3[i] > v_3[i+1] and not(v_3[i+1]==0):
    for j in range(i, m+1, 1):
        if(v_3[j] > v_3[j+1]):
            v_3[j+1], v_3[j] = baseSwape(v_3[j+1], v_3[j])
        else:
            break

elif v_3[i] < v_3[i-1]:
    for j in range(i, -1, -1):
        if(v_3[j] < v_3[j-1]):
            v_3[j-1], v_3[j] = baseSwape(v_3[j-1], v_3[j])
        else:
            break
print(v_3)

```

#### e) Procedimento de Intercalação Lista Ordenada

```python
print(v_4)

j=0
while(v_4[j] < v_4[j+1] and j<len(v_4)):
    j+=1

v_aux = [0]*len(v_4)

i=0
j+=1
mid = j
k=0

while(k<len(v_4)):
    if j>=len(v_4) or (i<mid and v_4[i]<v_4[j]):
        v_aux[k] = v_4[i]
        i += 1
    else:
        v_aux[k] = v_4[j]
        j += 1
    k+=1
    
print(v_aux)
    
```
