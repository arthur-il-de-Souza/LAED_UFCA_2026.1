# Trabalhando com Listas Ordenadas

## 1. Introdução

### Funções:

```python
def findM(lista: list[int]):
    m=0
    for i in lista:
        if not(i==0):
            m+=1
            continue
        break
    return m-1


def baseSwape( a: int, b: int):
    aux = a
    a = b
    b = aux

    return a, b
```

### Listas:

```python
v_1 = [3, 9, 2, 3, 2, 7, 3, 9, 5, 0, 0, 0] # Repetida/Desordenada
v_2 = [2, 2, 3, 3, 3, 5, 7, 9, 9, 0, 0, 0] # Repetido/Ordenado 
v_3 = [1, 3, 7, 8, 10, 16, 19, 21, 28, 0, 0, 0]

v_v = [5, 3, 9, 7, 8, 10, 13, 1, 2, 14, 21, 11]
v_u = [4, 8, 2, 11, 6, 5, 15, 3, 19, 1, 7, 19]
```

#### a) Números Repetidos em uma Lista Desordenada

```python
m = findM(v_1)
print(f"pos: {m}, num: {v_1[m]}")

for i in range(0, len(v_1), 1):
    for j in range(i+1, len(v_1), 1):
        if(v_1[i] == v_1[j]):
            v_1[j] = v_1[m+1]

m = findM(v_1)
print(f"pos: {m}, num: {v_1[m]}")

for i in range(0, len(v_1)-1, 1):
    m = findM(v_1)
    m+=1
    for j in range(m, len(v_1), 1):
        if v_1[j]==0:
            continue
        v_1[m], v_1[j] = baseSwape(v_1[m], v_1[j])
        break

for i in range(len(v_1)):
    print(f"{i}: {v_1[i]}")
        
```
