## 1. Método Input

O método input sempre, por padrão, recebe um valor do tipo  string [ 'str' ] sendo necessário uma conversão de tipo explícita para a alocação de outras variáveis.

```python
x = 5
y = input("Digite um número: ")
print(f"Como podemos ver, variáveis numéricas escritas sem aspas duplas são consideradas ({x}) do tipo {type(x)}")
print(f"input {y} do tipo {type(y)}")
```

## 2. Operações

### 2.1 Operações Algébricas

#### 2.1.1 Soma

```python
x = 10
y = 3

print(f"{x} + {y} = {x+y}")
```

#### 2.1.2 Subtração

```python
print(f"{x} - {y} = {x-y}")
```

#### 2.1.3 Multiplicação

```python
print(f"{x} * {y} = {x*y}")
```

#### 2.1.4 Divisão

```python
print(f"{x} / {y} = {x/y}")
```

#### 2.1.5 Inteiro da Divisão

```python
print(f"{x} // {y} = {x//y}")
```

#### 2.1.6 Resto da Divisão

```python
print(f"{x} % {y} = {x%y}")
```

### 2.2 Operações Lógicas

#### 2.2.1 Operação $>$ e $>=$

```python
print(f"{x} > {y} = {x>y}")
print(f"{y} >= {x} = {y>=x}")
```

#### 2.2.2 Operação $ < $ e $ <= $

```python
print(f"{x} < {y} = {x<y}")
print(f"{y} <= {x} = {y<=x}")
```

#### 2.2.3 Modificadores de operações **and, or e not**

```python
# == e modificadores and, or e not 

print(f"{x} == {y} = {x==y}")
print(f"({x} == {y}) or not({y} == {x}) = {(x==y) or (not(y>=x))}")
print(f"({x} == {y}) or not({y} == {x}) = {(x==y) and (not(y>=x))}")
print(f"not({y} == {x}) = {not(y>= x)}")
```

## 3. Estruturas Condicionais 

    if(validacao) -> funciona como uma cabeça para a estrutura condicional será, executado caso as operações lógicas resultarem em True, caso contrário o interpretador pulará esta etapa;

    elif(validacao) -> sucede um if ou outro elif, é executado após a estrutura condicional anterior ter retornado False e o resultado de suas operações lógicas resultarem em True, caso contrário o interpretador pulará esta etapa;

    else -> é utilizado no final de uma estrutura condicional, é executado quando nenhuma outra estrutura obteve um valor True dentro de seus parâmetros 

```python
if(y>x):
    print("1")
elif(y<x):
    print("2")
else:
    print("3")
```

## 4. Estruturas de Repetição

##### 4.1 For

A instância for é expressa da seguinte forma:

    for i in lista:
onde i vai representar sequêncialmente cada elemento da lista uma vez cada.

##### 4.2 While
A instância wihle é expressa da seguinte forma:

    while(validação):
Essa estrutura é executada enquanto o conteúdo da validação continuar sendo True, caso contrário ela terminará a execução desse trecho de código.
