#### EXERCÍCIO 1
Lista[1] Cria um lista de String com 10 posições. Preenche todas as posições com a palavra “id”

```python
lista = ["id"] * 10

print(len(lista))
cont = 0
for i in lista:
    cont+=1
    print(f"{cont}: {i}")
```

#### EXERCÍCIO 2
Lista [2]

Cria uma lista de int com 100 posições. Preenche cada posição com o seu número (por exemplo, na posição 0, deve estar o número 0). Imprime o valor contido em todas as posições, uma por linha.

```python
lista = [0] * 100

cont = 0
for i in range(len(lista)):
    lista[i] = cont
    cont+=1

print(len(lista))

for i in lista:
    print(i,end=" ")
```

#### EXERCÍCIO 5
Dias importantes

É necessário guardar 3 datas de aniversário no mês de dezembro. A Ana faz anos no dia 1, o Xavier faz anos no dia 10 e o João faz anos no dia 23. Cria uma lista com uma posição para cada dia do mês de dezembro. Adiciona os nomes dos aniversariantes na lista, na posição do seu dia de aniversário. No final, imprime o conteúdo de cada posição.

```python
dezembro = [""]*31

a1 = [1, 10, 23]
names = ["Ana", "Xavier", "João"]

cont = 0
for i in range(0, 31, 1):
    if (i == a1[cont]-1):
        dezembro[i] = names[cont]
        if(len(a1)-1 > cont):
            cont+=1
    print(f"dia {i+1}: {dezembro[i]}")
```

#### EXERCÍCIO 6
Escala de Dor

Um enfermeiro está a tratar um doente mas precisa de saber se o tratamento está a fazer efeito, isto é, se a dor do doente está a diminuir. O enfermeiro utiliza a escala universal de dor, onde o doente indica, numa escala de 1 a 10, qual é o nível da sua dor. O enfermeiro vai registar, durante uma semana, qual o nível de dor do doente. No final do registo, o programa deverá indicar qual foi o nível de dor do 1º dia, qual o nível de dor do 7º dia e qual o nível médio de dor ao longo dos 7 dias. Utiliza uma lista no desenvolvimento do programa.

```python
lista = [0] * 7

for i in range(len(lista)):
    num = int(input(f"Dia {i+1}: Qual o nível de dor?: "))
    lista[i] = num

print(f"Dor no primeiro dia: {lista[0]}")
print(f"Dor no sétimo dia: {lista[len(lista)-1]}")
print(f"A média de dor durando 7 dias foi de: {sum(lista)/len(lista)}")
```

#### EXERCÍCIO 7

Troca

Recebe uma lista de 10 números. O programa deverá indicar se há trios e quais são (um trio é uma sequência de 3 números iguais).

```python
from random import randint

lista = [0]*10

for i in range(len(lista)):
    lista[i] = randint(0, 10)
    print(lista[i], end=" ")
print(" ")

for i in range(len(lista)-1):
    cont = 0
    for j in range(i+1, len(lista),1):
        if(lista[i] == lista[j]):
            cont+=1
        if(cont==2):
            print(f"{lista[i]} == {lista[j]}")
            print(f"Existe sim")
            print(cont)
            print("")
            break
```
