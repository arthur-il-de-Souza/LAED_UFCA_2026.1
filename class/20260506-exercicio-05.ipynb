#### 11. Desenvolva um programa que peça ao utilizador 2 números inteiros e calcule a soma de todos os números inteiros compreendidos entre esses números (com os próprios limites incluídos). O utilizador deve colocar inserir primeiro o menor número e depois o maior número. Por exemplo, se o utilizador introduzir 10 e 14 o resultado deve ser 10+11+12+13+14.

```python
n1 = int(input("Informe o 1º Número: "))
n2 = int(input("Informe o 2º Número: "))

soma = 0
for i in range(n1, n2+1, 1):
    soma += i
print(soma)
```

#### 12. Altere o programa anterior para que este funcione mesmo que o utilizador introduza primeiro o maior número.

```python
n1 = int(input("Informe o 1º Número: "))
n2 = int(input("Informe o 2º Número: "))

if(n1>n2):
    aux = n1
    n1 = n2
    n2 = aux

soma = 0
for i in range(n1, n2+1, 1):
    soma += i
print(soma)
```

#### 13. Crie um pequeno programa que apresente a primeira posição em que o caractere “A” surge numa string introduzida pelo utilizador, ou que, em caso de insucesso, que esse caractere não se encontra no texto introduzido.

```python
texto = input("Informe uma Palavra: ")

pos = 0
for letra in texto:
    if letra.upper() == "A":
        print(f"Posição real é: {pos+1} \nA posição relativa na string é: {pos}")
        break
    pos+=1
```

#### 14. Programe em Pyhton o código que permita calcular o factorial de um número N introduzido (por exemplo, 6! = 6 x 5 x 4 x 3 x 2 x 1).

```python
n = int(input("Informe um Número: "))

fatorial = range(1,n+1,1)
prod = 1
for i in fatorial:
    prod = i * prod
print(prod)
```

### 16. Apresente um programa que mostre os primeiros N termos da sequência de Fibonacci. Nessa sequência, cada elemento é igual à soma dos dois elementos imediatamente anteriores. Considere que a sequência começa com 0 e 1, ou seja:
    0 1 1 2 3 5 8 13 21 …

```python
n = int(input("Informe um Número: "))
lista = range(0,n,1)
atual = 0
prox = 1
for i in range(len(lista)):
    print(atual)
    aux = atual + prox
    atual = prox
    prox = aux
```

#### 18. Crie um programa para verificar se o número que um utilizador introduz é primo (um número primo é um número natural maior do que 1 que só é divisível por 1 e por ele próprio).

```python
n = int(input("Informe um número: "))
cont = 0
for i in range(1, n+1, 1):
    if (n%i == 0):
        cont +=1
    if cont==3: break

if cont==2: print(f"{n} é primo!!")
if cont==3: print(f"{n} não é primo!!")
```
