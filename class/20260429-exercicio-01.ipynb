#### 7. Avaliação de notas: O programa pede 3 notas ao utilizador. Se alguma nota estiver abaixo da nota mínima (7.5 valores), ou se a média das notas estiver abaixo de 9.5, o aluno reprova. Senão, aprova.

```python
rep = 0
soma = 0
quantidade = 3

for i in range(quantidade):
    x = float(input(f"Informe a {i+1}ª nota: "))
    if(x<7.5):
        rep = 1
        break
    soma += x
soma /= quantidade
if(soma<9.5 or rep == 1):
    print(f"Reprovado!!")
else:
    print(f"Aprovado!! Média: {soma:2.2f}")

    
```

#### 8. Validador de triângulo retângulo: o utilizador insere o comprimento de 3 lados de um triângulo. O programa indica se o triângulo é retângulo ou não.

```python
from math import pow

lados = [0,0,0]
lados[0] = float(input("Informe o tamanho do Cateto Oposto: "))
lados[1] = float(input("Informe o tamnaho do Cateto Adjacente: "))
lados[2] = float(input("Informe o tamanho do Hipotenuza: "))

if(lados[0] + lados[1] <= lados[2]):
    print(f"O comprimento dos lados não forma um triângulo: {lados}")
elif((pow(lados[0],2)+pow(lados[1],2))==pow(lados[2],2)):
    print(f"O triângulo de lados {lados} é um triângulo retângulo")
else:
    print(f"Os lados {lados} não formam um triângulo retângulo")
```

#### 9. Pedindo ao utilizador um ponto (2 coordenadas) e o raio de uma circunferência de centro O = (0,0), o programa indica se o ponto inserido está dentro da circunferência central com o raio indicado.

```python
from math import sqrt, pow

cord = [0,0]
ponto = [0,0]

r = float(input("Informe o raio da circunferência: "))

for i in range(len(cord)):
    ponto[i] = float(input(f"Informe a {i+1} coordeanda: "))

if(r<sqrt(pow(ponto[0],2)+pow(ponto[1],2))):
    print(f"O ponto {ponto} não está contido em uma circunferência de raio {r} centrado na origem.")
else:
    print(f"O ponto {ponto} está contido em uma circunferência de raio {r} centrado na origem.")

```
