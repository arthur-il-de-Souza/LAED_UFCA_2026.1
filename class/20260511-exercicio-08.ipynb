### Fundamentos de Programação
#### Crie um programa que:
1. Solicite dois números.
2. Tente dividir o primeiro pelo segundo.
3. Trate os erros de:
1. valor inválido (ValueError)
2. divisão por zero (ZeroDivisionError)
4. Mostre o resultado apenas se não houver erro.
* utilize else e finally

```python
try:
    num1 = float(input("Informe o 1° Número: "))
    num2 = float(input("Informe o 2° Número: "))

    div = num1/num2
    print(f"Divisão:\n({num1}/{num2}) = {div}")
except ValueError:
    print("Informe um número Real!")
except ZeroDivisionError:
    print(f"{num2} não pode ser 0!")
finally:
    print(f"Acabou :)")
```

## Atividade Prática 2
### Fundamentos de Programação
#### Crie um programa que:
1. Solicite dois números.
2. Tente dividir o primeiro pelo segundo.
3. Trate os erros de:
1. valor inválido (ValueError)
2. divisão por zero (ZeroDivisionError)
4. Mostre o resultado apenas se não houver erro.
5. Repita até que as entradas sejam válidas. Para sair do programa o utilizador deve digitar a
“Sair” quando for solicitado.

```python
try:
    while True: 
        num1 = float(input("Informe o 1° Número: "))
        num2 = float(input("Informe o 2° Número: "))

        div = num1/num2
        print(f"Divisão:\n({num1}/{num2}) = {div}")

        val = input("\nDeseja continuar? (Sair p/ sair do programa)")

        if val.upper() == "SAIR":
            break

except ValueError:
    print("\nInforme um número Real!")
except ZeroDivisionError:
    print(f"\n{num2} não pode ser 0!")
finally:
    print(f"\nAcabou :)")
```
