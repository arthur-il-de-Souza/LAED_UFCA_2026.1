#### 2. Soma de números de 1 a 100: Usando um loop while, escreva um programa que calcula a soma dos números de 1 a 100.

```python
i = 0
soma = 0
while (i <= 100):
    soma += i
    i+=1
print(soma)
```

#### 4. Pular múltiplos de 3: Usando while e continue, escreva um programa que exibe números de 1 a 20, mas pula os múltiplos de 3.

```python
i = 0
while (i <= 20):
    if(i%3==0 and i !=0):
        i+=1
        continue
    print(i)
    i += 1
```

```python
i = 0

while(True):
    if(not (i%3==0 and i!=0)):
        print(i)
    if(i>20):
        break
    i += 1
```

#### 15. Apresente um programa que mostre os primeiros N termos da sequência de Fibonacci. Nessa sequência, cada elemento é igual à soma dos dois elementos imediatamente anteriores. Considere que a sequência começa com 0 e 1, ou seja:

```python
num = int(input("Informe o valor: "))
i = 0
j=[0,1]
print(j[0])
print(j[1])
while(i<=num):
    i+=1
    aux = j[1]
    j[1] = j[0]+j[1]
    j[0] = aux
    print(j[1])
```

#### 17. Crie um programa para verificar se o número que um utilizador introduz é primo (um número primo é um número natural maior do que 1 que só é divisível por 1 e por ele próprio).

```python
num = int(input("Informe um número: "))
i=1
cont = 0
while(i<=num):
    if(num%i==0):
        print(f"{num} é divisível por {i}")
        cont +=1
    i+=1
if(cont == 2):
    print(f"\n{num} é primo.", )
else:
    print(f"\n{num} não é primo!!!")
```

```python
num = int(input("Informe um número: "))
i=1
cont = 0
while(True):
    if(num%i==0):
        print(f"{num} é divisível por {i}")
        cont +=1
    i+=1
    if(i>num):
        break
if(cont == 2):
    print(f"\n{num} é primo.", )
else:
    print(f"\n{num} não é primo!!!")
```
