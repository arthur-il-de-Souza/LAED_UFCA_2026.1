# Questões 3 e 6

Funções utilzadas na resolução das questões:

1. __list_gen__: Para gerar uma lista de $n$ elementos dentro de um intervalo fechado $[a,b]$;
2. __swap__: troca dois elementos de posição;
3. __bubble_sort__: algoritmo de ordenação quadrático;
4. __bubble_sort_n__: algoritmo de ordenação quadrático que ordena elementos dentro de um intervalo $[a,b)$;
5. __particiona__: algoritmo que particiona uma lista levando o 1° elemento como o pivô;
6. __is_sorted__: verifica se a lista está completamente ordenada.

```python
import random

# Função que gera uma lista de n elementos com números do intervalo [a,b]
def list_gen(n: int, a: int, b: int): 
    c_list = [0]*n
    for i in range(0, n, 1):
        c_list[i] = random.randint(a, b)
    return c_list

# Troca a posição de dois elementos
def swap(a: int, b: int): 
    return b,a

# Algoritmo de Ordenação Quadrático
def bubble_sort(c_list: list[int]): 
    tam = len(c_list)
    for i in range(0, tam, 1):
        for j in range(0, tam-(i+1), 1):
            if(c_list[j]>c_list[j+1]):
                c_list[j],c_list[j+1] = swap(c_list[j],c_list[j+1])

    return c_list

# Algoritmo de Ordenação Quadrático dado um intervalo 
def bubble_sort_n(c_list: list[int], inicio: int, fim: int):
    tam = fim - inicio
    for i in range(tam):
        for j in range(inicio, fim-(i+1), 1):
            if(c_list[j]>c_list[j+1]):
                c_list[j],c_list[j+1] = swap(c_list[j],c_list[j+1])
    return c_list

# Algoritmo que particona a lista tendo como pivô o 1° elemento
def particiona(c_list= list[int], inicio= int, fim= int):
    x = c_list[inicio]
    i = inicio-1
    for j in range(inicio+1, fim):
        if c_list[j]<x:
            i+=1
            c_list[j], c_list[i] = swap(c_list[j], c_list[i])
    c_list[inicio], c_list[i] = swap(c_list[inicio], c_list[i])
    return c_list, i + 1

# Busca Binária
def b_search(k: int, c_list: list[int], inicio: int, fim: int):
    mid = int((inicio+fim)/2)

    if(c_list[mid] == k):
        return mid
    
    if(c_list[mid]>k and inicio<=fim):
        return b_search(k, c_list, inicio, mid-1)
    
    elif(c_list[mid]<k) and inicio<=fim:
        return b_search(k, c_list, mid+1, fim)
        

# Algoritmo que verifica se a lista está ordenada
def is_sorted(c_list: list[int]):
    for i in range(0, len(c_list)-1, 1):
        if c_list[i] > c_list[i+1]:
            return -1
    return 0
```

### 3. Imagine que nós ordenamos os primeiros $\frac{2n}{3}$ elementos da lista $L$ usando um algoritmo qualquer com tempo de execução $\mathcal{O}(n^2)$
### Daí, nós ordenamos os últimos $\frac{2n}{3}$ elementos da lista usando o mesmo algoritmo de ordenação quadrático.
### Finalmente, nós ordenamos os primeiros $\frac{2n}{3}$ elementos da lista outra vez usando o algoritmo

#### Lista instanciada

1. $n$: tamanho da lista;
2. $a,b$: representa o intervalo que dos números que serão gerados;
3. $lista$: lista de elementos retornada pela função __list_gen__

```python
n = 12 # Tamanho da lista
a,b = 0, 10 # Intervalo fechado dos valores alocados na lista
lista = list_gen(n, a, b) # Gera uma lista de n elemetos, onde cada valor é um número inteiro dentro do intervalo [a,b]
print(lista)
```

```python
inicio = 0
fim = int((2*n)/3)
lista[inicio: fim] = bubble_sort_n(lista, inicio, fim)[inicio: fim] # Ordena os elementos do intervalo [0, (2*n)/3]

inicio = int(n/3)
fim = n
lista[inicio: fim] = bubble_sort_n(lista, inicio, fim)[inicio: fim] # Ordena os elementos do intervalo [n/3, n]

inicio = 0
fim = int((2*n)/3)
lista[inicio: fim] = bubble_sort_n(lista, inicio, fim)[inicio: fim] # Ordena os elementos do intervalo [0, (2*n)/3]

print(lista)
print(is_sorted(lista)) # Verifica se a lista está completamente ordenada, retorna 0 (lista ordenada) ou  -1 (lista não ordenada)
```

#### a) Ao final desse processo, a lista está completamente ordenada? Porque?

Resposta = 

Vamos verificar a execução do código em cada trecho da ordenação.

1. Serão ordenados os $\frac{2n}{3}$ primeiros números;

2. Serão ordenados os números da posição $\frac{n}{3}$ até o $n$-ésimo valor;

3. Por fim, serão ordenados, novamente, os $\frac{2n}{3}$ primeiros números.

Ao final do processo podemos verificar que a interseção entre as listas, cujo intervalo é $[\frac{n}{3}, \frac{2n}{3}]$ fará a lista ser completamente ordenada se a quantidade elementos que perteceria ao intervalo $[0,\frac{2n}{3}]$, mas estão no intervalo $[\frac{n}{3}, n]$, ou $[\frac{n}{3}, n]$, mas estão no intervalo $[0,\frac{2n}{3}]$ seja igual ou menor a amplitude da interseção, ou seja $\frac{n}{3} $


#### b) Analise o tempo de execução desse procedimento.
1. Na primeira parte do algoritmo os $\frac{2n}{3}$ primeiros números serão ordenados por um algoritmo quadrático, portanto $\mathcal{O}(n^2)$;
2. Na sequência, serão ordenados, também por meio de um algortmo de ordenação quadrático, os s números da posição $\frac{n}{3}$ até o $n$-ésimo valor, portanto $\mathcal{O}(n^2)$;
3. Por fim, retornaremos ao passo 1., que nos dá um tempo quadrático, portanto $\mathcal{O}(n^2)$.


### 6. Imagine que nós aplicamos o procedimento Partição() sobre a lista $L$ em relação ao seu primeiro elemento $ x $
### E imagine que depois nós ordenamos os dois lados da partição usando um algoritmo qualquer com tempo de execução $ \mathcal{O}(n^2) $

```python
n = 12 # Número de Elementos contidos na lista
a, b = 0, 1000 # Intervalo em que os números serão gerados

lista = list_gen(n, a, b) # Função que cria a lista
print(lista)
```

```python
lista, pos = particiona(lista, 0, n) # Recebe a lista particionada e a posição

lista[0: pos] = bubble_sort_n(lista, 0, pos)[0: pos]
lista[pos: n] = bubble_sort_n(lista, pos, n)[pos: n]

print(lista)
print(is_sorted(lista))
```

#### a) Analise o tempo de execução desse algoritmo no melhor caso.
O melhor caso acontece quando o pivô (primeiro elemento) é o elemento que divide a lista ao meio, quando executado o algoritmo __particiona__.
1. A primeira metade da lista é organizada em $\mathcal{O}(\frac{n^2}{4})$;
2. A segunda metade da lista é organizada também em $\mathcal{O}(\frac{n^2}{4})$;
3. Portanto, a espreção na notação Big O é dada por: $(\frac{n^2}{4}) + (\frac{n^2}{4}) = \frac{n^2}{2}$

O que resulta em um tempo: $\mathcal{O}(\frac{n^2}{2})$

#### b) Analise o tempo de execução desse algoritmo no pior caso.
O pior caso acontece quando o pivô é o primeiro elemento ou o último elemento na lista, quando executado o algoritmo __particiona__.
1. O pivô, sendo o último ou o primeiro elemento, só haverá uma parte para ser ordenada, composta por $n-1$ elementos;
2. Organizando os $n-1$ elementos restantes temos um tempo dado pela espressão: $$(n-1)^2 = n^2 -2n +1$$

O que resulta em um tempo quadrático: $\mathcal{O}(n^2)$
