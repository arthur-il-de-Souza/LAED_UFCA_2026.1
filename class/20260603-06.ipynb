```python
import random

# Função que gera uma lista de n elementos com números do intervalo [a,b]
def list_gen(n: int, a: int, b: int): 
    c_list = [0]*n
    for i in range(0, n, 1):
        c_list[i] = random.randint(a, b)
    return c_list

# Troca a posição de dois elementos
def swap(a: int, b: int): 
    return b,a

# Algoritmo de Ordenação Quadrático
def bubble_sort(c_list: list[int]): 
    tam = len(c_list)
    for i in range(0, tam, 1):
        for j in range(0, tam-(i+1), 1):
            if(c_list[j]>c_list[j+1]):
                c_list[j],c_list[j+1] = swap(c_list[j],c_list[j+1])

    return c_list

# Algoritmo de Ordenação Quadrático dado um intervalo 
def bubble_sort_n(c_list: list[int], inicio: int, fim: int):
    tam = fim - inicio
    for i in range(tam):
        for j in range(inicio, fim-(i+1), 1):
            if(c_list[j]>c_list[j+1]):
                c_list[j],c_list[j+1] = swap(c_list[j],c_list[j+1])
    return c_list

# Algoritmo que particona a lista tendo como pivô o 1° elemento
def particiona(c_list= list[int], inicio= int, fim= int):
    x = c_list[inicio]
    i = inicio-1
    for j in range(inicio+1, fim):
        if c_list[j]<x:
            i+=1
            c_list[j], c_list[i] = swap(c_list[j], c_list[i])
    c_list[inicio], c_list[i] = swap(c_list[inicio], c_list[i])
    return c_list, i + 1

# Busca Binária
def b_search(k: int, c_list: list[int], inicio: int, fim: int):
    mid = int((inicio+fim)/2)

    if(c_list[mid] == k):
        return mid
    
    if(c_list[mid]>k and inicio<=fim):
        return b_search(k, c_list, inicio, mid-1)
    
    elif(c_list[mid]<k) and inicio<=fim:
        return b_search(k, c_list, mid+1, fim)
        

# Algoritmo que verifica se a lista está ordenada
def is_sorted(c_list: list[int]):
    for i in range(0, len(c_list)-1, 1):
        if c_list[i] > c_list[i+1]:
            return -1
    return 0
```

### 3. Imagine que nós ordenamos os primeiros $\frac{2n}{3}$ elementos da lista L usando um algoritmo qualquer com tempo de execução $\mathcal{O}(n^2)$
### Daí, nós ordenamos os últimos $\frac{2n}{3}$ elementos da lista usando o mesmo algoritmo de ordenação quadrático.
### Finalmente, nós ordenamos os primeiros $\frac{2n}{3}$ elementos da lista outra vez usando o algoritmo

```python
n = 1_000 # Tamanho da lista
a,b = 0, 10 # Intervalo fechado dos valores alocados na lista
lista = list_gen(n, a, b) # Gera uma lista de n elemetos, onde cada valor é um número inteiro dentro do intervalo [a,b]
print(lista)
```

```python
inicio = 0
fim = int((2*n)/3)
lista[inicio: fim] = bubble_sort_n(lista, inicio, fim)[inicio: fim] # Ordena os elementos do intervalo [0, (2*n)/3]

inicio = int(n/3)
fim = n
lista[inicio: fim] = bubble_sort_n(lista, inicio, fim)[inicio: fim] # Ordena os elementos do intervalo [n/3, n]

inicio = 0
fim = int((2*n)/3)
lista[inicio: fim] = bubble_sort_n(lista, inicio, fim)[inicio: fim] # Ordena os elementos do intervalo [0, (2*n)/3]

print(lista)
print(is_sorted(lista)) # Verifica se a lista está completamente ordenada, retorna 0 (lista ordenada) ou  -1 (lista não ordenada)
```

#### a) Ao final desse processo, a lista está completamente ordenada? Porque?

Resposta = 

Vamos verificar a execução do código em cada trecho da ordenação.

i) Serão ordenados os $\frac{2n}{3}$ primeiros números;

ii) Serão ordenados os números da posição $\frac{n}{3}$ até o n-ésimo valor;

iii) Por fim, serão ordenados, novamente, os $\frac{2n}{3}$ primeiros números.

Ao final do processo podemos verificar que a interseção entre as listas, cujo intervalo é $[\frac{n}{3}, \frac{2n}{3}]$ fará a lista ser completamente ordenada se a quantidade elementos que perteceria ao intervalo $[0,\frac{2n}{3}]$, mas estão no intervalo $[\frac{n}{3}, n]$, ou $[\frac{n}{3}, n]$, mas estão no intervalo $[0,\frac{2n}{3}]$ seja igual ou menor a amplitude da interseção, ou seja $\frac{n}{3} $
